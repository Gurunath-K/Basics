```python
import os
import numpy as np
import scipy as sc
import skimage as sk
import matplotlib.pyplot as plt
from scipy import misc

```

```python
folder = 'F:\Amrita\Sem2\Images_Assignment3'
count=0
xt= []
result = np.zeros((25,25)) 

mat1 = np.zeros((10,1875))
mat2 = np.zeros((10,625))

def rgb2bin(rgb, thresh):
    u, v = rgb.shape
    for i in range(0,u):
        for j in range(0,v):
            if rgb[i][j]> thresh:
                result[i][j] = 255
            else:
                result[i][j] = 0
    return result
    
        
def rgbtogray(rgb):    
    fil= [0.299, 0.587, 0.114]
    return np.dot(rgb , fil)


for filename in os.listdir(folder): # iterate images in the folder.
    img = sc.misc.imread(os.path.join(folder,filename))
    thresh= sk.filters.threshold_mean(img)
    plt.imshow(img) #original image
    plt.title('Original')
    plt.show()
    
    res_img = sc.misc.imresize(img,(25,25))
    
    plt.imshow(res_img) #resized image
    plt.title('Resized')
    plt.show()
    grsc = rgbtogray(res_img)
    blk_wht = rgb2bin(grsc,thresh)
      
    plt.imshow(blk_wht, cmap = 'gray') #binary image
    
    plt.title('Binary')
    plt.show()
    xt += [res_img]
    x= res_img.reshape(1,1875) #25X25X3 vector
    y= blk_wht.reshape(1,625) #25X25 vector
    
    
    
    mat1[count,:] = np.array(x) 
    mat2[count,:] = np.array(y) 
    count+=1
    
    
print(mat1)
print(mat2)

print(mat1.shape)
print(mat2.shape)
```
