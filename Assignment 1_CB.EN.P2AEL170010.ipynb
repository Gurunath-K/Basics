```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
A = pd.read_csv('train.csv')
A.shape
M=A.as_matrix()
total=np.zeros(784)
total = total.reshape(28,28)
for i in range(10):
    im = M[i,1:]
    im = im.reshape(28,28)
    total+=im
    plt.imshow(im, cmap='gray')
    plt.show()

total= total/10    
plt.imshow(total, cmap='gray')
plt.show()
```
